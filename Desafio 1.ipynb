```python
#Importando bibliotecas

import cv2
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import heapq

%matplotlib inline
```

```python
#Utilizando função criada pelo professor para exibir o histograma de cores baseado no modelo CIECAM02

def histogram_hsv(image_hsv):
    hsv_image = cv2.cvtColor(image_hsv, cv2.COLOR_BGR2HSV)
    hue_hist = cv2.calcHist([hsv_image], [0], None, [180], [0, 180])
    plt.plot(hue_hist)
    plt.show()
```

```python
#Carregando as 3 imagens do desafio

image1 = cv2.imread('imagens/futebol.jpg')
image2 = cv2.imread('imagens/moda.jpg')
image3 = cv2.imread('imagens/simpsons.jpeg')
```

```python
#Primeiro convertemos a imagem de BGR para RGB, plotamos na tela, fazemos o cálculo do Histograma HSV para achar
#o valor na escala Hue e selecionamos os 3 maiores valores (heapq.nlargest). Por fim mostramos na tela os 3 valores
#e plotamos o histograma abaixo para melhor visualizar os resultados. Repetimos o processo para as próximas 2
#imagens para confirmar que o cálculo da cor na escala Hue está correto.

showimage1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,7))
plt.imshow(showimage1, cmap="gray")
plt.show()
hsv1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
hue1 = cv2.calcHist([hsv1], [0], None, [180], [0, 180])

posit = heapq.nlargest(3, range(len(hue1)), hue1.take)

print('Top 3 cores mais predominantes de acordo com posição Hue do modelo CIECAM02: ', posit)
histogram_hsv(image1)
```

```python
showimage2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,7))
plt.imshow(showimage2, cmap="gray")
plt.show()
hsv2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)
hue2 = cv2.calcHist([hsv2], [0], None, [180], [0, 180])

posit = heapq.nlargest(3, range(len(hue2)), hue2.take)

print('Top 3 cores mais predominantes de acordo com posição Hue do modelo CIECAM02: ', posit)
histogram_hsv(image2)
```

```python
showimage3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,7))
plt.imshow(showimage3, cmap="gray")
plt.show()
hsv3 = cv2.cvtColor(image3, cv2.COLOR_BGR2HSV)
hue3 = cv2.calcHist([hsv3], [0], None, [180], [0, 180])

posit = heapq.nlargest(3, range(len(hue3)), hue3.take)

print('Top 3 cores mais predominantes de acordo com posição Hue do modelo CIECAM02: ', posit)
histogram_hsv(image3)
```
